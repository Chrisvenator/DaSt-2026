1. Introduction
For this project we ended up using five different metadata standards to describe different parts of our ML experiment. At first it felt like a lot of overhead, but going through each one made it clear that they actually cover very different things. This document looks at all ten pairs of standards and tries to honestly identify where they overlap, where they diverge, and where we noticed actual conflicts in our own files.
