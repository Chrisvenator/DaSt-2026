# T3.11 – Metadata Standards Overlap Analysis

## Project: DaSt-2026 – Road Traffic Accident Severity Classification

Author: Sravanthi Muthineni (Owner C)

Date: 26 May 2026

# 1. Introduction

In this project, we used five different metadata standards to describe different parts of our machine learning experiment. At first, using so many standards felt a bit unnecessary, but after comparing them properly, it became clear that each one focuses on a different aspect of the project.

This analysis examines all ten pairs of standards and discusses where they overlap, where they differ, and whether any inconsistencies appeared across our metadata files. The discussion is based on the actual files created in our repository rather than only theoretical definitions.

## Metadata Standards Used

| Standard | File | Purpose |
|---|---|---|
| RO-Crate 1.1 | `ro-crate-metadata.json` | Describes the whole experiment package,  including files, datasets, people, and relationships |
| CodeMeta 2.0 | `codemeta.json` | Describes software metadata and dependencies, runtime environment, authors, and licence |
| FAIR4ML | `docs/fair4ml_severity_rf_baseline_v1.json` | Describes the ML model and evaluation |
| Croissant (MLCommons) | `docs/20260521_croissant.json` | Describes dataset structure and fields,units, and distributions |
| Model Card | `docs/model-card.md` | Human-readable model documentation |

# 2. Pairwise Analysis

## 2.1 RO-Crate ↔ CodeMeta

### Shared Information

Both standards describe the software itself. They contain the software name, authors with ORCIDs, licence information, version, Zenodo DOI(10.5281/zenodo.20182653), and GitHub repository URL.

### What RO-Crate Adds

RO-Crate connects the software to other artefacts in the project. It links scripts, datasets, models, and outputs together and includes provenance relationships between them.
It also includes both TUWRD deposit DOIs (10.70124/3ykwc-3sg80 for the model deposit and 10.70124/5zb61-c7k62 for the generated data deposit) and the DBRepo database entity.
### What CodeMeta Adds

CodeMeta focuses more on software environment details such as dependencies, Python runtime version, operating system compatibility, issue tracker, and README URL.

### Conflicts Noticed

No conflicts in the final versions of the files. Both consistently use MIT as the code licence. During the project we did notice an earlier version of ro-crate-metadata.json used GPL-3.0 for the code files while codemeta.json used MIT — this was corrected before the final submission and both files now agree on MIT.

## 2.2 RO-Crate ↔ FAIR4ML

### Shared Information

Both standards reference the trained model artefact, authors, dataset UUID, CC BY 4.0 licence, Model Card, and Zenodo DOI.

### What RO-Crate Adds

RO-Crate describes all output files individually and records relationships between datasets, scripts, and outputs. It also includes links to external dataset providers.

### What FAIR4ML Adds

FAIR4ML contains machine learning specific information such as hyperparameters, preprocessing configuration, dataset split statistics, evaluation metrics, intended use, and limitations.

### Conflicts Noticed

No conflicts were identified.The UUIDs, ORCIDs, and licence values were consistent between both files. However, if the model is retrained in the future, both files would need to be updated together.

## 2.3 RO-Crate ↔ Croissant

### Shared Information

Both standards describe the dataset using the dataset name, licence, DBRepo URL, and references to data.europa.eu. Both also reference dataset distribution endpoints.

### What RO-Crate Adds

RO-Crate focuses on provenance and relationships between datasets, scripts, and outputs. It also describes organisations involved in dataset creation and publication.

### What Croissant Adds

Croissant provides detailed column-level dataset information. It defines all 25 fields, their data types, units, and whether fields are categorical.

### Conflicts Noticed

The creator and publisher information is written slightly differently between the files. The meanings are compatible, but the wording could be aligned better.

## 2.4 RO-Crate ↔ Model Card

### Shared Information

Both standards describe the model name, dataset reference, licence information, authors, and related references.

### What RO-Crate Adds

RO-Crate includes machine-readable graph structure, checksums, encoding formats, file sizes, and DOI relationships.

### What the Model Card Adds

The Model Card provides readable explanations about intended use, limitations, ethics, evaluation results, and dataset bias discussions.

### Conflicts Noticed

The evaluation metrics currently match between both files, but future retraining could easily introduce inconsistencies if both files are not updated together.

## 2.5 CodeMeta ↔ FAIR4ML

### Shared Information

Both standards include author information, project name, description, version, licence information, DOI, and GitHub repository links.

### What CodeMeta Adds

CodeMeta provides software environment details such as dependencies, runtime version, operating system support, issue tracker, and README links.

### What FAIR4ML Adds

FAIR4ML focuses on machine learning specific metadata including algorithms, hyperparameters, preprocessing, evaluation metrics, and reproducibility information.

### Conflicts Noticed

The licences describe different artefacts. CodeMeta uses MIT for software, while FAIR4ML uses CC BY 4.0 for the trained model artefact. This distinction is intentional and correct — they cover different artefact types — but it needs to remain clearly scoped to avoid confusion

## 2.6 CodeMeta ↔ Croissant

### Shared Information

Only a few generic fields overlap, such as  name, description, licence, URL, and identifier.

### Differences

CodeMeta focuses on software metadata, while Croissant focuses on dataset structure and field-level metadata.

### Conflicts Noticed

No conflicts were found between these two standards.

## 2.7 CodeMeta ↔ Model Card

### Shared Information

Both include author information, project name, version, and licence details.

### What CodeMeta Adds

CodeMeta includes software environment information such as dependencies and runtime details.

### What the Model Card Adds

The Model Card focuses on model behaviour, evaluation results, intended use, ethical considerations, and limitations.

### Conflicts Noticed

No conflicts were identified. The licence distinction is clear and intentional — MIT for the code (CodeMeta) and CC BY 4.0 for the model output (Model Card). Both licences are confirmed and finalised in the repository.

## 2.8 FAIR4ML ↔ Croissant

### Shared Information

Both standards reference the same training dataset, DBRepo UUID, dataset URL, licence, temporal coverage, and dataset endpoint.

### What FAIR4ML Adds

FAIR4ML focuses on the trained model, including algorithms, hyperparameters, evaluation metrics, and dataset split statistics.

### What Croissant Adds

Croissant provides detailed dataset field definitions, unit URIs, distribution endpoints, and column-level metadata.

### Conflicts Noticed

The dataset endpoint URLs matched correctly between both files, so no conflicts were identified.

## 2.9 FAIR4ML ↔ Model Card

### Shared Information

This pair has the highest overlap. Both standards describe model information, intended use, limitations, ethical considerations, dataset references, and evaluation metrics.

### What FAIR4ML Adds

FAIR4ML provides machine-readable metadata including reproducibility details, hyperparameters, and model artefact information.

### What the Model Card Adds

The Model Card explains the same information in a readable narrative form and discusses issues such as dataset bias and misleading accuracy interpretation.

### Conflicts Noticed

Currently, all metric values match because both files were generated from the same `metrics.json` output. However, this pair is at the highest risk of future inconsistencies if the model is ever retrained.

## 2.10 Croissant ↔ Model Card

### Shared Information

Both standards reference the same training dataset, DBRepo UUID, dataset URL, licence, temporal coverage, and spatial coverage.

### Differences

Croissant focuses on dataset structure, while the Model Card focuses on model evaluation and behaviour.

### Conflicts Noticed

No conflicts were identified.

# 4. Discussion

## 4.1 Are the Standards Complementary or Redundant?

Overall, the standards are mostly complementary rather than redundant. Each one focuses on a different part of the project. Looking back, we probably would not have chosen five separate standards if it was up to us, but each one ended up covering something the others missed.

## 4.2 Where is Synchronisation Most Difficult?

The hardest pair to keep synchronised is FAIR4ML and the Model Card because they share a large amount of overlapping information in different formats.

During the project we noticed an earlier version of RO-Crate used GPL-3.0 while CodeMeta and CITATION.cff used MIT. This was corrected before final submission and all files now consistently use MIT for the code licence.

## 4.3 What is Missing from All Standards?

One thing we noticed is that the ontology mappings created in T2.2 are not represented in any of the five metadata standards.

Another missing area is API pagination details. None of the standards document how data was retrieved using paginated REST API requests. A user would still need to read the Python loader script or README documentation to fully reproduce the data loading process.
